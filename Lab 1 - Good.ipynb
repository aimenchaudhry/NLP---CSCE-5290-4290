```python
#Aimen Chaudhry 11326859 CSCE 4290 Team member: Kalyan Bandaru 

from bs4 import BeautifulSoup

import urllib.request

import nltk
#urllib module to crawl the web page
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX') #grab a web page content  

html = response.read()

soup = BeautifulSoup(html,"html.parser") #to clean the grabbed text

text = soup.get_text(strip=True)

tokens = [t for t in text.split()] #convert text into tokens 

freq = nltk.FreqDist(tokens) #calculate the frequency distribution 

for key,val in freq.items():
    print (str(key) + ':' + str(val))
    
```

```python

from nltk.corpus import stopwords 
stopwords.words('english') #To get English stop words
```

```python
clean_tokens = tokens [:]

sr = stopwords.words('english')

for token in tokens: # iterate over the tokens and remove the stop words
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
```

```python
freq = nltk.FreqDist(clean_tokens)
```

```python
freq.plot(10, cumulative=False) #plot a graph for those tokens using plot function 
```
